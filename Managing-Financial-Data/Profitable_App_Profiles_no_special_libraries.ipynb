```python
from csv import reader
```

```python
#opening AppleStore.csv file and storing data as a list of lists
opened_a = open("AppleStore.csv")
from csv import reader
apple_list = list(reader(opened_a))
```

```python
#opening googleplaystore.csv file and storing data as a list of lists
opened_g = open("googleplaystore.csv")
from csv import reader
google_list = list(reader(opened_g))
```

```python
def explore_data(dataset):
    """Function takes in one parameter, which is expected to be a list of lists
       Function prints several rows from the data set.
       Function prints the number of rows and columns."""
    
    dataset_slice = dataset[0:5]
    for data in dataset_slice:
        print(data)
        print('\n')
    print('Number of rows:',len(dataset))
    print('Number of columns:', len(dataset[0]))
```

```python
#explore Apple dataset
explore_data(apple_list)
```

```python
#explore google dataset
explore_data(google_list)
```
